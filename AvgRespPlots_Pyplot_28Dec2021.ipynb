# Normalized average responses as linear saliency

## Description

Comparing the average response estimates in each condition (High Threat, Low Threat, High Reward and Low Reward) with their normalized counterparts (mean across conditions subtracted from average response of each condition). Reward outcome types (received, missed) are averaged in each condition. Error bands are Â±1 SE around mean.

```python
import numpy as np
import matplotlib.pyplot as plt
import csv
import math
#import matplotlib as mpl
#mpl.rcParams['figure.dpi'] = 300
```

## Conditions

```python
conditions=[('HT'), ('LT'), ('HR'), ('LR')]
```

## Reading the files

```python
mAvg = []
seAvg = []
mNormAvg = []
seNormAvg = []

for cond in range(4):
    
    file = open('mSub_AvgResp_' + conditions[cond] +'.csv')
    csvreader = csv.reader(file,quoting=csv.QUOTE_NONNUMERIC)
    arr = []
    for row in csvreader:
        arr.append(row)
    file.close()
    mAvg.append(arr)
    
    file = open('seSub_AvgResp_' + conditions[cond] +'.csv')
    csvreader = csv.reader(file,quoting=csv.QUOTE_NONNUMERIC)
    arr = []
    for row in csvreader:
        arr.append(row)
    file.close()
    seAvg.append(arr)
    
    file = open('mSub_NormAvgResp_' + conditions[cond] +'.csv')
    csvreader = csv.reader(file,quoting=csv.QUOTE_NONNUMERIC)
    arr = []
    for row in csvreader:
        arr.append(row)
    file.close()
    mNormAvg.append(arr)
    
    file = open('seSub_NormAvgResp_' + conditions[cond] +'.csv')
    csvreader = csv.reader(file,quoting=csv.QUOTE_NONNUMERIC)
    arr = []
    for row in csvreader:
        arr.append(row)
    file.close()
    seNormAvg.append(arr)
```

### Reading ROI names

```python
file = open('ROI_Names.csv')
csvreader = csv.reader(file)
ROIname = []
for row in csvreader:
        ROIname.append(row)
file.close()
```

# Plotting

*Legend*: Similar colour code as before

**Dull** colours are **Average** Response Estimates, 
**Bright** colours are **Normalized** Response estimates 

## High Threat

```python
# HIDE CODE

mAvgResp = np.array(mAvg)[0]
seAvgResp = np.array(seAvg)[0]
mNormAvgResp = np.array(mNormAvg)[0]
seNormAvgResp = np.array(seNormAvg)[0]

fig, axs = plt.subplots(nrows=17, ncols=5, figsize=(20, 30))

for idx in np.arange(85):
    ax = axs[idx//5, np.mod(idx,5)]
    ax.title.set_text(ROIname[idx])
    ax.plot(np.zeros(13), c='black')
    ax.plot(np.squeeze(mAvgResp[:, idx]), c='#ff8c00')
    ax.plot(np.squeeze(mNormAvgResp[:, idx]), c='#ff4500')
    ax.fill_between(np.arange(13), np.squeeze(mAvgResp[:, idx]-seAvgResp[:, idx]), np.squeeze(mAvgResp[:, idx]+seAvgResp[:, idx]), alpha=0.5, facecolor='#ff8c00')
    ax.fill_between(np.arange(13), np.squeeze(mNormAvgResp[:, idx]-seNormAvgResp[:, idx]), np.squeeze(mNormAvgResp[:, idx]+seNormAvgResp[:, idx]), alpha=0.5, facecolor='#ff4500')

plt.subplots_adjust(left=0.1, bottom=0.1, right=0.9, top=0.9, wspace=0.3, hspace=0.5)
```
